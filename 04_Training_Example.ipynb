```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

# Training Example

Let's train the model $y=ax+b$.

## Dataset

```python
class linear_data():
    
    def __init__(self, a, b, min_x=0.0, max_x=10.0):
        self.a = a
        self.b = b
        self.min_x = min_x
        self.max_x = max_x
    
    def get_true_data(self, batch_size):
        x = np.random.uniform(self.min_x,self.max_x,(batch_size))
        y = self.a*x + self.b
        return x.reshape((batch_size,1)), y.reshape((batch_size,1))
    
    def get_data(self, batch_size):
        x = np.random.uniform(self.min_x,self.max_x,(batch_size))
        y = self.a*x + self.b + np.random.uniform(-2*self.a, 2*self.a, (batch_size))
        return x.reshape((batch_size,1)), y.reshape((batch_size,1))
```

```python
data = linear_data(25.0, 12.5)
x, y = data.get_data(100)
plt.scatter(x, y)
```

## Placeholders

```python
x = tf.placeholder(tf.float32,(None, 1))
y = tf.placeholder(tf.float32,(None, 1))
```

## Variables

```python
a = tf.Variable(tf.random_normal([1]))
b = tf.Variable(tf.zeros([1]))
```

## Operations (graph)

```python
y_pred = a*x + b
error = tf.reduce_sum(tf.square(y-y_pred))
```

## Optimizer

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.00001)
train = optimizer.minimize(error)
```

## Training

```python
batch_size = 100
iterations = 1000
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    for i in range(iterations):
        x_data, y_data = data.get_data(batch_size)
        print(x_data.shape, y_data.shape)
        sess.run(train,feed_dict={x:x_data, y:y_data})
        print(sess.run(error,feed_dict={x:x_data, y:y_data}))
    new_a = sess.run(a)
    new_b = sess.run(b)
```

```python
x_data, y_data = data.get_data(100)
plt.scatter(x_data, y_data)
plt.scatter(x_data, x_data*new_a[0] + new_b[0])
```
